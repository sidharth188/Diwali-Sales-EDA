```python
import numpy as  np # we are importing numpy which is use for working with array and save it in np
import pandas as pd # we are importing pandas which is use for data manipulation and cleaning and save it in pd
import matplotlib.pyplot as plt # we are importing matplotlib.pyplot which is use for making graphs and save it in plt
%matplotlib inline
import seaborn as sns # we are importing seaborn which is use for making plots and save it in sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape') # importing csv file
```

```python
df.shape # to check numners of rows and column
```

```python
df.head(10) # it will show 10 datas from dataset
```

```python
df.info() # it wiill show column and if there is any null value important for data cleaning
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True) #deleteing two blank clolumn will have null values with axis=1 which select full row of that two column
```

```python
pd.isnull(df) #to check null values
```

```python
pd.isnull(df).sum() # this will show how maany null values
```

```python
df.dropna(inplace=True)# delete null values
```

```python
df['Amount'] = df['Amount'].astype('int')# change datatypes of any cloumn
```

```python
df.columns#shows all datatypes
```

```python
df.rename(columns= {'Marital_Status':'Shadi'})#changle column name
```

```python
df.describe()#give you details of dataset
```

```python
ax=sns.countplot(x='Gender',  data=df) #it will plot count graph 
```

```python
ax=sns.countplot(x='Gender',  data=df) #it will plot count graph 
for bars in ax.containers:
    ax.bar_label(bars) # now it will also show data
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)#showing data in numbers
```

```python
a=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data= a)
print("most of the shppping done by females")
```
